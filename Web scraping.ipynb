```python
import urllib3
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
warnings.filterwarnings('ignore')
```

```python
url='https://www.yelp.com/search?find_desc=Sandwiches&find_loc=Boston%2C+MA+2108&ns=1'
```

```python
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
response.data
```

```python
soup=BeautifulSoup(response.data, "lxml")
print(soup.prettify())
```
