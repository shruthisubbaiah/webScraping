```python
import urllib3
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
warnings.filterwarnings('ignore')
url='https://in.bookmyshow.com/buytickets/jumanji-welcome-to-the-jungle-3d-bengaluru/movie-bang-ET00067940-MT/20180119'
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
bs=BeautifulSoup(response.data, "lxml")
print(bs.prettify())
```

```python
multiplex=bs.find_all("div",class_="listing-info")
for m in multiplex:
    if m.div["class"]==["details"]:
        print "yes"
```

```python
multiplex_names=[]

multiplex=bs.find_all("div",class_="__name")
for m in multiplex:
        multiplex_names.append(m.strong.get_text())
        
multiplex_names
```

```python
movie=bs.find("div",class_="cinema-name-wrapper")
movie_name=movie.a.get_text()
print movie_name
```

```python
genre_list=[]
genre=bs.find_all("span",class_="__genre-tag")
for g in genre:
    genre_list.append(g.get_text())
genre_list
```
